```python

import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import os,os.path
import gensim,re
import operator
```

```python
class Sentence(object):
    def __init__(self, directory):
        self.directory = directory
 
    def __iter__(self):
        for each in os.listdir(self.directory):
            if each != '.DS_Store':
                
                with open(os.path.join(self.directory,each,'sentences.txt'),'r', encoding='utf-8') as obj:
                    for each_line in obj:
                        review_text = re.sub("[^a-zA-Z]"," ", each_line)
                        yield review_text.split()
```

```python
sent = Sentence('C:/Users/Prithika/nlp project/stackexchange')
```

```python
from collections import defaultdict
d = defaultdict(lambda : 0)
for i in sent:
    d[len(i)] += 1
```

```python
d = dict(sorted(d.items(), key=operator.itemgetter(1), reverse=True)[:10])
```

```python
model = gensim.models.Word2Vec(sent,  workers=4, size=300, min_count = 1, window = 15, sample = 1e-3)
```

```python

mname_saved = "Word2vec_model"
model.save(mname_saved)
```

```python
print(len(model.wv.vocab))
```

```python
print(model.similarity('pen','android'))
print(model.similarity('wallet','purse'))
print(model.similarity('windows','microsoft'))
```
